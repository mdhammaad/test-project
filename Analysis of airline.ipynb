```python
# Total flight by airline
data['airline'].unique()
```

```python
data['airline'].value_counts(ascending=True).plot(kind='barh', color =('lightblue','skyblue','blue','mediumblue', 'darkblue','navy'))
plt.title("Total Flights per Airline")
plt.xlabel("Number of Flights")
plt.ylabel("Airline")
plt.show()
```

```python
# Total flight by depurture and arrival
data['departure_time'].value_counts(ascending=True).plot(kind='bar', color =('g','y'))
plt.title("Total Flights per Departure Time")
plt.xlabel("Number of Flights")
plt.ylabel("Departure Time")
plt.show()
data['arrival_time'].value_counts(ascending=True).plot(kind='bar', color =('g','y'))
plt.title("Total Flights per Arrival Time")
plt.xlabel("Number of Flights")
plt.ylabel("Arrival Time")
plt.show()
```

```python
# Total flight by cities
data.head()
```

```python
data['source_city'].value_counts() 
```

```python
data['destination_city'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.barh(data['source_city'].value_counts().index, data['source_city'].value_counts().values, color=('r','b'))
plt.title("Total Flight Source City")
plt.xlabel("Number of Flights")
plt.ylabel("Source City")

plt.subplot(1,2,2)
plt.barh(data['destination_city'].value_counts().index, data['destination_city'].value_counts().values, color=('r','b'))
plt.title("Total Flight Destination City")
plt.xlabel("Number of Flights")
plt.ylabel("Destination City")
plt.show()
```

```python
## Average price by airline
data.head()
```

```python
data.groupby('airline')["price"].mean().sort_values(ascending=True)
```

```python
sb.catplot( x = 'airline' , y = 'price' , kind = 'bar' , palette = 'rocket' , data = data, hue = 'airline' )
plt.title("Average Price per Airline")
sb.catplot( x = 'airline' , y = 'price' , kind = 'bar' , palette = 'rocket' , data = data, hue = 'class' )
plt.title("Average Price per Airline by Class")
plt.show()
```

```python
# Average price by timing
data.head()
```

```python
data.groupby(["departure_time"])["price"].mean().sort_values(ascending=True)
```

```python
data.groupby(["arrival_time"])["price"].mean().sort_values(ascending=True)
```

```python
plt.figure(figsize=(10,6))
sb.catplot( x = 'departure_time' , y = 'price' , kind = 'bar' , palette = 'dark' , data = data, hue = 'departure_time' )
plt.title("Average Price per Departure Time")
plt.figure(figsize=(10,6))
sb.catplot( x = 'arrival_time' , y = 'price' , kind = 'bar' , palette = 'dark' , data = data , hue = 'arrival_time' )
plt.title("Average Price per Arrival Time")
plt.show()
```

```python
sb.relplot( x = 'arrival_time' , y = 'price' , data = data , col = 'departure_time' , kind = 'line' )

plt.show()
```

```python
# Average price by day left
data.head()
```

```python
filtered_data = data[data['days_left'].isin([1,2])]
sb.barplot(x='days_left', y='price', data=filtered_data, palette='magma', hue='days_left')
plt.title('Average Price vs Days Left (1-2 days)'), plt.xlabel('Days Left Before Flight'), plt.ylabel('Average Price')
plt.show()
```
